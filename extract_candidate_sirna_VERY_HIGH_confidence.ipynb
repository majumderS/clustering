```python
#extract window
def extract_list_of_sequences(fasta,chromosome,start,end,window_length):
    list_sequences = []
    seq = ''
    for record in SeqIO.parse(fasta,"fasta"):
        if(chromosome in record.description):
            seq = str(record.seq[start:end])
            print(record)
            break
    for i in range(0,len(seq)-window_length):
        seq1 = seq[i:i+window_length]
        seq1_c =str(Seq(seq[i:i+window_length]).reverse_complement())
        list_sequences.append((seq1,seq1_c))
    return list_sequences
```

```python
def get_gc_percent(forward_strand,reverse_strand):
    total_nt = len(forward_strand) + len(reverse_strand)
    gc_count = 0
    for chr in list(forward_strand):
        if(chr=='g' or chr=='c'):
            gc_count = gc_count + 1
    for chr in list(reverse_strand):
            if(chr=='g' or chr=='c'):
                gc_count = gc_count + 1
    # print(float(gc_count/total_nt)*100.0)
    return (float(gc_count/total_nt)*100.0)
```

```python
fasta = "Human_genomes/human_genome_full.fasta"
win_len = 24
forward_strand = []
reverse_strand = []
chrid = []
gene_A = []
gene_B = []
coexpression_score = []
confidence_class = []
overlap_length = []
gc_percent = []
for idx,row in filtered_coexp_data.iterrows():
    L = extract_list_of_sequences(fasta,row['chr_id'],row['overlap_start'],row['overlap_end'],win_len)
    for item in L:
        forward_strand.append(item[0].upper())
        reverse_strand.append(item[1].upper())
        gc_percent.append(get_gc_percent(item[0],item[1]))
        chrid.append(row['chr_id'])
        gene_A.append(row['gene_A'])
        gene_B.append(row['gene_B'])
        coexpression_score.append(row['coexpression_score'])
        confidence_class.append(row['confidence_class'])
        overlap_length.append(row['overlap_length'])
    
    print("DONE-------------->",row['chr_id'],row['gene_A'],row['gene_B'])


table_of_extracted_duplex = pd.DataFrame({'Forward_strand':forward_strand,'Reverse_strand':reverse_strand,'gc_percent':gc_percent,\
             'chr_id':chrid,'gene_A':gene_A,'gene_B':gene_B,'coexpression_score':coexpression_score,\
             'confidence_class':confidence_class,'overlap_length':overlap_length})
        
    
```

```python
df1 = error_free_data1[['ganterior','gmiddle','gposterior']]
df2 = error_free_data2[['ganterior','gmiddle','gposterior']]
df3 = error_free_data3[['ganterior','gmiddle','gposterior']]
frames = [df1,df2,df3]
consolidated_dataframe = pd.concat(frames)
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
model = KMeans(n_clusters=4)
labels= model.fit_predict(consolidated_dataframe[['ganterior','gmiddle','gposterior']])
# Plotting the clusters
# fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6),pro)

fig = plt.figure(figsize = (20, 20))
ax = plt.axes(projection ="3d")
# K-means plot
ax.scatter3D(consolidated_dataframe['ganterior'],consolidated_dataframe['gposterior'], consolidated_dataframe['gmiddle'],c=labels, cmap='viridis')
ax.set_title('Clustering of duplex_core, duple_start and duplex_end values',fontsize=20)
ax.set_xlabel('Ganterior',fontsize=20)
ax.set_ylabel('Gposterior',fontsize=20)
ax.set_zlabel('Gmiddle',fontsize=20)

plt.savefig('clustering.png')
 
```

```python
# Elbow Method for Optimal Cluster Number
distortions = []
K = range(1, 20)  # Test clusters from 1 to 10

for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(consolidated_dataframe[['ganterior', 'gmiddle', 'gposterior']])
    distortions.append(kmeanModel.inertia_)

# Plotting the Elbow plot
plt.figure(figsize=(8, 6))
plt.plot(K, distortions, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```
